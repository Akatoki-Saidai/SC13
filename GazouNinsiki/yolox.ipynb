```python
from google.colab import drive
drive.mount('/content/drive')

%cd ./drive/MyDrive
#!git clone https://github.com/Megvii-BaseDetection/YOLOX


#ディレクトリー構成
"""

|-drive
    |
    |-Mydrive
        |
        |-etc...
        |
        |-YOLOX
        |   |
        |   |-etc...
        |   |
        |   |-yolox
        |   |
        |   |-ダウンロードした重みのファイル
        |   |
        |   |-学習モデルが入ったフォルダー(YOLOX-outputs)
        |   |
        |   |-学習モデルを読み込み、物体検出するプログラムを書いたコード
        |
        |
        |-dataset   
            |
            |-annotations(cocoフォーマットのアノテーションデータを入れる)
            |
            |-train2017(学習画像)
            |
            |-val2017(評価画像)

基本的に学習させる際は、train,val,testに画像をわけるが、YOLOXはtrain,valしか指定させることができなかったので、フォルダーを作ってない
＊追記 self.test_ann = とすれば追加できるかも

違う画像やサイズで学習させたい場合は
/content/drive/MyDrive/YOLOX/exps/example/custom/nano.py
を開いて、input_size を変えたり、dataset path を変える。(nanoやsの場合はcustomでいいけど、それより大きいモデルを用いる場合はdefaultにあるのを使う)
けど、できるだけフォルダーの名前は上記の構成で用いた名前にする。(書き換える必要があるところが増えて面倒)

自前のPCでやる際は学習環境の構築のためにCUDAやcuDNNをインストールしたり、ライブラリーのバージョンや依存関係が面倒なのでおすすめしない

学習する際にYOLOX/tools/demo.pyが呼び出されて、そこで from yolox.data.datasets import COCO_CLASSES
とあるが、COCO_CLASSESに検知する物体の名前を書いておく必要がある。(不必要なものは消してもいい)
もしくは自分で物体の名前を書いたのを作り、それをdemo.pyのCOCO_CLASSESを呼び出しているところと置き換えてあげてもいいと思う。

*追記　画像のサイズは416*416にしたほうがいい。じゃないと何かしらエラーが出る
*追記　使わせていただくプログラム自体がダウンロードした時期によって、不具合を含んでいることがあるので、その場合は自分で対処か過去のをダウンロードする。
*追記　クラス数を1つにするとエラーになったりするかも
"""
```
